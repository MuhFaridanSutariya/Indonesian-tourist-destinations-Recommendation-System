## Improve imdb score

```python
class RecommenderSystem:
    def __init__(self, data):
        self.df = pd.read_csv(data)
    
    def recommend(self, Category=None, City=None, Price=None, topk=10):
        df = self.df.copy()
        df = self.demographic_filter(df, Category=Category, City=City, Price=Price)
        
        result = df.sort_values("count", ascending=False)
        result = result.head(topk) 
        return result
    
    @staticmethod
    def demographic_filter(df, Category=None, City=None, Price=None, topk=10):
        df = df.copy()
        
        if Category is not None :
            df = df[(df['Category'] == Category)]
        if City is not None :
            df = df[(df['City'] == City)]
        if Price is not None :
            df = df[(df['Price'].between(Price[0], Price[1]))]
        return df
```

```python
recsys = RecommenderSystem(data='data/df_tourism.csv')
```

```python
recsys.recommend(Category='Culture', City='Semarang')
```
